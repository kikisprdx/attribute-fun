In this guided demonstration, we'll explore how machine learning models can inadvertently leak information about sensitive attributes in the training data. This type of privacy leakage is known as an "attribute inference attack."

## What is an Attribute Inference Attack?

An attribute inference attack aims to infer sensitive attributes about individuals whose data was used to train a machine learning model. The attacker assumes that they have access to:
1. Some non-sensitive attributes of an individual
2. Either the model outputs (black-box) or the model parameters (white-box)


The goal is to predict sensitive attributes that were intentionally excluded from the released data.

## The Nursery Dataset

I'll demonstrate these attacks on the Nursery dataset, where the sensitive attribute is the 'social' feature. In the original dataset, this is a categorical feature with 3 values: 'problematic', 'slightly_prob', and 'nonprob'. To simplify, I've hot-end encoded this into a binary feature where 'problematic' = 1 and the other values = 0.




````python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import torch
import seaborn as sns
from torch.utils.data import DataLoader

from src.data.loader import NurseryDataLoader
from src.models.classifier import NurseryClassifier, ModelTrainer
from src.attacks.black_box import AttributeInferenceBlackBox
from src.attacks.white_box import AttributeInferenceWhiteBox
from src.attacks.baseline import AttributeInferenceBaseline
from src.utils.evaluation import evaluate_attack
from src.utils.visualization import (
    plot_feature_importance,
    visualize_information_leakage,
    visualize_attack_confidence,
    plot_confusion_matrix,
    compare_attack_performance
)


np.random.seed(42)
torch.manual_seed(42)
````

````python
print("Loading the Nursery dataset...")
data_loader = NurseryDataLoader(test_size=0.5, transform_social=True)
(x_train, y_train), (x_test, y_test), feature_names, social_indices, (train_dataset, test_dataset) = data_loader.load()

print(f"Training data shape: {x_train.shape}")
print(f"Testing data shape: {x_test.shape}")
print(f"Social feature indices: {social_indices}")

attack_feature = social_indices[0]
````

## Target Model Training

Before conducting attribute inference attacks, we need to establish a Target Model. This model is trained on non-sensitive attributes to perform a legitimate prediction task. In our case, we're using a neural network classifier trained on the Nursery dataset.

The Target Model consists of:
- Input layer matching the dimensionality of non-sensitive features
- Hidden layer with 64 neurons
- Output layer corresponding to the main classification task

This model represents a standard machine learning deployment scenario where sensitive attributes have been intentionally excluded from the training process for privacy reasons.

Despite these precautions, we'll demonstrate how information about sensitive attributes might still be encoded within the model's parameters or revealed through its predictions.

````python
print("\nTraining the target model...")
input_size = x_train.shape[1]  # Number of features
hidden_size = 64
num_classes = len(np.unique(y_train))

# Train the model
model = NurseryClassifier(input_size, hidden_size, num_classes)
trainer = ModelTrainer(model)
model = trainer.train(train_dataset, num_epochs=10, batch_size=64, verbose=True)

# Evaluate the model
accuracy = trainer.evaluate(test_dataset)
print(f'Target model accuracy: {accuracy:.4f}')
````

**For the attacks, we'll split our training data into two parts:**

2. Data for evaluating the attack models

````python
print("\nPreparing data for the attacks...")
(attack_x_train, attack_x_train_feature, attack_x_train_without_feature,
 attack_x_test, attack_x_test_feature, attack_x_test_without_feature,
 unique_values) = data_loader.prepare_attack_data(attack_feature, train_ratio=0.5)

print(f"Attack training data shape: {attack_x_train.shape}")
print(f"Attack testing data shape: {attack_x_test.shape}")
print(f"Unique values of the attack feature: {unique_values}")
````

````python
def create_confusion_matrix(model, test_dataset, feature_names=None):
    test_loader = DataLoader(dataset=test_dataset, batch_size=64, shuffle=False)
    all_predictions = []
    all_labels = []
    with torch.no_grad():
        for inputs, labels in test_loader:
            outputs = model(inputs)
            _, predicted = torch.max(outputs.data, 1)
            _, labels_max = torch.max(labels.data, 1)
            all_predictions.extend(predicted.cpu().numpy())
            all_labels.extend(labels_max.cpu().numpy())
    return all_predictions, all_labels

nursery_input_features = [
    'parents',      # usual, pretentious, great_pret
    'has_nurs',     # proper, less_proper, improper, critical, very_crit
    'form',         # complete, completed, incomplete, foster
    'children',     # 1, 2, 3, more
    'housing',      # convenient, less_conv, critical
    'finance',      # convenient, inconv
    'social',       # nonprob, slightly_prob, problematic
    'health'        # recommended, priority, not_recom
]

nursery_class_names = [
    'Not Recommended',   # 0
    'Priority',          # 1
    'Special Case',      # 2 
    'Very Recommended',  # 3
    'Recommended'        # 4
]

print("\nTraining Features Used:")
for idx, name in enumerate(nursery_input_features):
    print(f"{idx+1}. {name}")

# Get da labels
predictions, true_labels = create_confusion_matrix(model, test_dataset)
unique_classes = np.unique(np.concatenate([np.unique(true_labels), np.unique(predictions)]))

print("\nTarget Classes (Output Classes):")
for idx, name in enumerate(nursery_class_names):
    status = "✓" if idx in unique_classes else "✗"
    print(f"Class {idx}: {name} {status}")

print(f"\nUnique classes in true labels: {np.unique(true_labels)}")
print(f"Unique classes in predictions: {np.unique(predictions)}")
print(f"Total instances evaluated: {len(true_labels)}")

class_names = [nursery_class_names[i] for i in unique_classes]

conf_matrix = pd.DataFrame(
    np.zeros((len(unique_classes), len(unique_classes))),
    index=class_names,
    columns=class_names
)

for pred, true in zip(predictions, true_labels):
    # Find the correct positions based on unique_classes
    true_idx = np.where(unique_classes == true)[0][0]
    pred_idx = np.where(unique_classes == pred)[0][0]
    conf_matrix.iloc[true_idx, pred_idx] += 1

# Class-wise accuracy
print("\nClass-wise accuracy:")
for i, class_idx in enumerate(unique_classes):
    indices = np.where(np.array(true_labels) == class_idx)[0]
    if len(indices) > 0:
        correct = np.sum(np.array(predictions)[indices] == class_idx)
        accuracy = correct / len(indices)
        print(f"{nursery_class_names[class_idx]}: {accuracy:.4f} ({correct}/{len(indices)})")

# Normalize confusion matrix
norm_conf_matrix = conf_matrix.div(conf_matrix.sum(axis=1), axis=0).fillna(0)

# Plot confusion matrix
plt.figure(figsize=(12, 10))
sns.set(font_scale=1.2)
heatmap = sns.heatmap(norm_conf_matrix, annot=True, fmt='.2f', cmap='Blues',
                    cbar_kws={'label': 'Normalized Frequency'})
heatmap.set_title('Target Model Performance: Normalized Confusion Matrix')
heatmap.set_xlabel('Predicted Class')
heatmap.set_ylabel('True Class')
plt.tight_layout()
plt.show()
````

````python
# Split the training data for the attack
attack_train_ratio = 0.5
attack_train_size = int(len(x_train) * attack_train_ratio)
attack_test_size = int(len(x_train) * attack_train_ratio)
attack_x_train = x_train[:attack_train_size]
attack_y_train = y_train[:attack_train_size]
attack_x_test = x_train[attack_train_size:]
attack_y_test = y_train[attack_train_size:]

print(f"Attacking feature at index: nonprob")

# We seperate the data into two sections:
# One without the sensitive feature (what an attacker would have)
# One with the sensitive feature (What the attacker wants to learn, test dataset)
x_test_tensor = torch.FloatTensor(attack_x_test)
with torch.no_grad():
    attack_x_test_predictions = model(x_test_tensor).numpy()
attack_x_test_without_feature = np.delete(attack_x_test, attack_feature, 1)
attack_x_test_feature = attack_x_test[:, attack_feature].copy().reshape(-1, 1)

# Black-box model, trained on the same inputs/outputs of the source model
bb_attack = AttributeInferenceBlackBox(model, attack_feature=attack_feature)
bb_attack.fit(attack_x_train)

# Get the unique values of the attacked feature
unique_values = np.unique(attack_x_test_feature)
print(f"Unique values of the attacked feature: {unique_values}")

# Infer the attacked feature values
inferred_train_bb = bb_attack.infer(attack_x_test_without_feature, pred=attack_x_test_predictions, values=unique_values)

# Check accuracy
train_acc = np.sum(inferred_train_bb == attack_x_test_feature.reshape(1, -1)) / len(inferred_train_bb.ravel())
print(f"Black-box attack accuracy: {train_acc:.4f}")
````

## White-box Attack Implementation

We now implement the most powerful form of attribute inference attack: the **white-box attack**. Unlike the black-box attack which only uses model outputs, the white-box attack leverages complete access to the model's internal parameters, weights, and gradients.

### How White-box Attacks Work

White-box attacks operate by exploiting the model's decision boundaries and gradient information to recover sensitive attributes. They work through gradient-based optimization:

1. For each test instance (with the sensitive attribute removed), the attack tries to determine what value of the sensitive attribute would "fit" best with the model's internal representation
2. The attack uses gradient information to iteratively update its guess of the sensitive attribute value
3. By analyzing how changing the sensitive attribute would affect the model's internal state, the attack can often accurately reconstruct the missing value

### Implementation Details

```python
# Initialize the white-box attack with the target model
wb_attack = AttributeInferenceWhiteBox(target_model=model, attack_feature=attack_feature)

# Infer the sensitive attribute through gradient-based optimization
inferred_wb = wb_attack.infer(
    attack_x_test_without_feature,
    values=unique_values,
    learning_rate=0.01,  # Controls step size for gradient updates
    iterations=50        # Number of optimization iterations
)
```

This attack runs an optimization process for each test instance, using the model's gradients to guide the search for the most likely value of the sensitive attribute.

### Attack Evaluation

We evaluate the effectiveness of the white-box attack using standard classification metrics:

```python
wb_results = evaluate_attack(
    inferred_wb,
    attack_x_test_feature,
    is_categorical=True,
    positive_value=unique_values[1] if len(unique_values) > 1 else unique_values[0]
)
```

The evaluation compares the inferred sensitive values against the ground truth to determine attack accuracy, precision, recall, F1-score, and AUC.

### Privacy Implications

White-box attacks represent the worst-case scenario for privacy leakage. If an adversary has complete access to model parameters:

1. Removing sensitive attributes from training data provides little privacy protection
2. The model's learned representations may contain recoverable sensitive information
3. Even carefully trained models might leak information through their internal parameter values

The effectiveness of white-box attacks demonstrates the need for stronger privacy-preserving techniques beyond simply excluding sensitive attributes from training data.

````python
# The white-box attack uses the target model's internal parameters and gradients
# to infer the sensitive attribute. This is the most powerful attack but requires
# complete access to the model.

wb_attack = AttributeInferenceWhiteBox(target_model=model, attack_feature=attack_feature)

# Infer the sensitive attribute
inferred_wb = wb_attack.infer(
    attack_x_test_without_feature,
    values=unique_values,
    learning_rate=0.01,
    iterations=50  
)

# Evaluate the attack
wb_results = evaluate_attack(
    inferred_wb,
    attack_x_test_feature,
    is_categorical=True,
    positive_value=unique_values[1] if len(unique_values) > 1 else unique_values[0]
)

print("White-box Attack Results:")
for metric, value in wb_results.items():
    print(f"  {metric}: {value:.4f}")
````

## Comparing Attack Performance

To understand the extent of information leakage, we compare three different attribute inference attacks with increasing levels of access to the target model.

### The Baseline Attack

The baseline attack serves as our control experiment:

```python
baseline_attack = AttributeInferenceBaseline(attack_feature=attack_feature)
baseline_attack.fit(attack_x_train)
inferred_baseline = baseline_attack.infer(attack_x_test_without_feature, values=unique_values)
```

This attack relies solely on statistical correlations within the non-sensitive features to infer the sensitive attribute. It doesn't use the target model at all, representing the minimum information an attacker could extract just from public data.

### Evaluation Metrics

We evaluate all three attacks using multiple metrics to get a comprehensive understanding of their effectiveness:

1. **Accuracy**: Overall percentage of correctly inferred sensitive values
2. **Precision**: Proportion of positive identifications that were correct
3. **Recall**: Proportion of actual positives that were correctly identified

For the 'social_problematic' attribute:
- Precision tells us: "When the attack predicts a family has problematic social conditions, how often is it right?"
- Recall tells us: "What percentage of families with problematic social conditions does the attack correctly identify?"

### Comparative Analysis

The comparative analysis reveals:

1. **If black-box > baseline**: The target model's predictions leak information about the sensitive attribute
2. **If white-box > black-box**: Additional information is leaked through the model's internal parameters
3. **Attack ranking**: Typically white-box > black-box > baseline, with the margin between them indicating how much additional information is leaked at each access level

### Privacy Implications

If the black-box or white-box attacks significantly outperform the baseline, this demonstrates that:

1. Simply removing sensitive attributes from training data is insufficient for privacy protection
2. Models can inadvertently memorize and leak sensitive information through their behavior
3. The more access an attacker has to the model, the more sensitive information they can extract

````python
baseline_attack = AttributeInferenceBaseline(attack_feature=attack_feature)
baseline_attack.fit(attack_x_train)
inferred_baseline = baseline_attack.infer(attack_x_test_without_feature, values=unique_values)
baseline_acc = np.sum(inferred_baseline == attack_x_test_feature.reshape(1, -1)) / len(inferred_baseline.ravel())
print(f"Baseline attack accuracy: {baseline_acc:.4f}")

# Use our already calculated black-box attack accuracy
print(f"Black-box attack accuracy: {train_acc:.4f}")

# White-box accuracy
wb_acc = np.sum(inferred_wb == attack_x_test_feature.reshape(1, -1)) / len(inferred_wb.ravel())
print(f"White-box attack accuracy: {wb_acc:.4f}")

# Calculate precision and recall for all attacks
positive_value = unique_values[1] if len(unique_values) > 1 else unique_values[0]

# Define a function to calculate precision and recall (same as in original script)
def calc_precision_recall(predicted, actual, positive_value):
    """
    Calculate precision and recall for binary classification.

    Args:
        predicted: Predicted values
        actual: Actual values
        positive_value: Value to consider as positive class

    Returns:
        tuple: (precision, recall)
    """
    # Ensure arrays are 1D
    if isinstance(predicted, np.ndarray) and predicted.ndim > 1:
        predicted = predicted.reshape(-1)
    if isinstance(actual, np.ndarray) and actual.ndim > 1:
        actual = actual.reshape(-1)

    # Calculate true positives, false positives, false negatives
    tp = np.sum((predicted == positive_value) & (actual == positive_value))
    fp = np.sum((predicted == positive_value) & (actual != positive_value))
    fn = np.sum((predicted != positive_value) & (actual == positive_value))

    # Calculate precision and recall
    precision = tp / (tp + fp) if (tp + fp) > 0 else 0
    recall = tp / (tp + fn) if (tp + fn) > 0 else 0

    return precision, recall

# Accuracy
baseline_precision, baseline_recall = calc_precision_recall(inferred_baseline, attack_x_test_feature, positive_value)
bb_precision, bb_recall = calc_precision_recall(inferred_train_bb, attack_x_test_feature, positive_value)
wb_precision, wb_recall = calc_precision_recall(inferred_wb, attack_x_test_feature, positive_value)

all_results = {
    "Baseline": {"accuracy": baseline_acc, "precision": baseline_precision, "recall": baseline_recall},
    "Black-box": {"accuracy": train_acc, "precision": bb_precision, "recall": bb_recall},
    "White-box": {"accuracy": wb_acc, "precision": wb_precision, "recall": wb_recall}
}

# Print detailed results
print("\nDetailed attack results:")
for attack_name, metrics in all_results.items():
    print(f"\n{attack_name} Attack:")
    for metric, value in metrics.items():
        if metric.title() == 'f1_score':
            continue
        print(f"  {metric.title()}: {value:.4f}")

# Use visualization function if available
comparison = compare_attack_performance(all_results)
````

````python
def plot_custom_feature_importance(attack_model, feature_names, attack_feature_idx):
    weights = attack_model.layer1.weight.data.numpy()

    # Get feature names without the attack feature
    remaining_features = np.delete(feature_names, attack_feature_idx)

    input_names = list(remaining_features)
    if len(input_names) < weights.shape[1]:
        input_names.append('Model Predictions')

    if len(input_names) != weights.shape[1]:
        if len(input_names) > weights.shape[1]:
            input_names = input_names[:weights.shape[1]]
        else:
            input_names.extend([f'Unknown {i}' for i in range(weights.shape[1] - len(input_names))])

    # Compute the importance as the mean absolute weight for each input feature
    importance = np.mean(np.abs(weights), axis=0)
    importance_df = pd.DataFrame({
        'Feature': input_names,
        'Importance': importance
    })

    # Sort by importance
    importance_df = importance_df.sort_values('Importance', ascending=False)
    plt.figure(figsize=(12, 8))
    sns.barplot(x='Importance', y='Feature', data=importance_df.head(20))
    plt.title('Feature Importance for Black-box Attribute Inference Attack')
    plt.tight_layout()
    plt.show()

plot_custom_feature_importance(bb_attack.attack_model, feature_names, attack_feature)
````

## Conclusion: Privacy Implications of Attribute Inference Attacks

In this notebook, I've demonstrated how machine learning models can inadvertently leak sensitive information through attribute inference attacks. Let's summarize our key findings and their implications.

### Experimental Overview

I've explored three increasingly powerful attack vectors against a neural network trained on the Nursery dataset:

1. **Baseline Attack**: Uses only non-sensitive attributes to infer the sensitive attribute (social_problematic), establishing a minimum leakage benchmark
2. **Black-box Attack**: Exploits model predictions as additional information, representing an attacker with query access
3. **White-box Attack**: Leverages model internals and gradients, representing an attacker with complete model access

Our target model was trained to predict nursery admission decisions while intentionally excluding the sensitive social attribute, mimicking real-world privacy-preserving practices.

### Key Findings

Our comparative analysis revealed several important insights:

1. **Information Leakage**: The black-box attack outperformed the baseline, demonstrating that the model leaks information about the sensitive attribute despite not being explicitly trained on it.

2. **Access Impact**: The white-box attack achieved the highest accuracy, confirming that deeper access to model internals enables more effective privacy attacks.

3. **Correlation Exploitation**: All attacks succeeded by exploiting statistical correlations between sensitive and non-sensitive attributes, with the model inadvertently encoding these correlations in its parameters and predictions.

### Privacy Implications

These findings have several important implications for machine learning privacy:

1. **Insufficient Protection**: Simply removing sensitive attributes from training data does not adequately protect privacy, as models can still leak this information through indirect correlations.

2. **Deployment Considerations**: Organizations deploying machine learning models should consider the risk of attribute inference attacks when determining access levels and deployment strategies.

3. **Beyond Simple Exclusion**: More sophisticated privacy-preserving techniques are needed, such as:
   - Differential privacy during training
   - Adversarial training to remove sensitive information from representations
   - Formal privacy guarantees rather than relying on security through obscurity

### Broader Impact

This experiment demonstrates that machine learning models can become unintentional vectors of privacy leakage. As ML systems are increasingly deployed in sensitive domains like healthcare, finance, and education, addressing these vulnerabilities becomes crucial.

By understanding how attribute inference attacks work and evaluating their effectiveness, I can develop more robust defenses and better protect sensitive information in the age of machine learning.
