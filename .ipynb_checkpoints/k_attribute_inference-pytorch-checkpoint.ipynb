```python
import os
import torch
import pandas as pd
import numpy as np
from urllib.request import urlretrieve
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.model_selection import train_test_split
from torch.utils.data import TensorDataset, DataLoader
import itertools

class NurseryDataset:
    """
    Class to load and preprocess the nursery dataset.
    """
    
    def __init__(self, test_set=0.5, transform_social=True, random_state=42):
        """
        Initialize and load the dataset.
        
        :param test_set: Proportion of the dataset to include in the test split
        :param transform_social: If True, transform the social feature to binary (0,1)
        :param random_state: Random seed for reproducibility
        """
        self.test_set = test_set
        self.transform_social = transform_social
        self.random_state = random_state
        self.column_names = ['parents', 'has_nurs', 'form', 'children', 'housing', 
                            'finance', 'social', 'health', 'class']
        self.class_names = None
        
        # Load and process the data
        self._download_data_if_needed()
        self._load_and_preprocess_data()
        
    def _download_data_if_needed(self):
        """Download the dataset if it doesn't exist locally."""
        self.dataset_path = 'nursery.data'
        if not os.path.exists(self.dataset_path):
            print("Downloading nursery dataset...")
            url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/nursery/nursery.data'
            urlretrieve(url, self.dataset_path)
    
    def _load_and_preprocess_data(self):
        """Load and preprocess the dataset."""
        # Load the dataset
        data = pd.read_csv(self.dataset_path, header=None, names=self.column_names)
        
        # Store original data
        self.raw_data = data.copy()
        
        # Transform the social feature to binary if required
        if self.transform_social:
            # Map 'problematic' to 1, everything else to 0
            social_mapping = {'problematic': 1, 'slightly_prob': 0, 'nonprob': 0}
            data['social'] = data['social'].map(social_mapping)
            self.social_values = [0, 1]  # The possible values after transformation
        
        # Extract the target variable (class) and store unique class names
        self.y = data['class']
        self.class_names = self.y.unique().tolist()
        self.X = data.drop('class', axis=1)
        
        # Get the social feature for later use
        self.social_feature = self.X['social']
        
        # One-hot encode categorical features
        categorical_features = self.X.columns
        self.encoder = OneHotEncoder(sparse_output=False, handle_unknown='ignore')
        self.X_encoded = self.encoder.fit_transform(self.X)
        
        # Scale the features
        self.scaler = StandardScaler()
        self.X_scaled = self.scaler.fit_transform(self.X_encoded)
        
        # Split the dataset
        self.x_train_np, self.x_test_np, self.y_train_np, self.y_test_np = train_test_split(
            self.X_scaled, self.y, test_size=self.test_set, random_state=self.random_state
        )
        
        # Get feature names after one-hot encoding
        self.feature_names = self.encoder.get_feature_names_out(categorical_features)
        
        # Find the index of the social feature in the transformed dataset
        self.social_feature_indices = [i for i, name in enumerate(self.feature_names) 
                                      if name.startswith('social_')]
        
        # Process target labels - get the actual one-hot encoded vectors
        y_dummies = pd.get_dummies(self.y)
        self.target_names = y_dummies.columns.tolist()
        
        # Convert to PyTorch tensors - IMPORTANT: store the actual dummies for train/test
        self.y_train_dummies = pd.get_dummies(self.y_train_np)
        self.y_test_dummies = pd.get_dummies(self.y_test_np)
        
        self.x_train = torch.FloatTensor(self.x_train_np)
        self.y_train = torch.FloatTensor(self.y_train_dummies.values)
        self.x_test = torch.FloatTensor(self.x_test_np)
        self.y_test = torch.FloatTensor(self.y_test_dummies.values)
        
        # Create PyTorch datasets
        self.train_dataset = TensorDataset(self.x_train, self.y_train)
        self.test_dataset = TensorDataset(self.x_test, self.y_test)
    
    def print_dataset_head(self, n=5):
        """
        Print the first n samples of the dataset with feature names and target labels.
        
        :param n: Number of samples to print
        """
        print(f"Dataset Head (first {n} samples):")
        print("=" * 80)
        
        # Get the first n samples from train_dataset
        samples = [self.train_dataset[i] for i in range(min(n, len(self.train_dataset)))]
        x_sample = torch.stack([x for x, _ in samples]).numpy()
        y_sample = torch.stack([y for _, y in samples]).numpy()
        
        # Convert to pandas DataFrame for easier printing
        features_df = pd.DataFrame(x_sample, columns=self.feature_names)
        
        # Create a DataFrame for targets with class names - this was the issue
        targets_df = pd.DataFrame(y_sample, columns=self.y_train_dummies.columns)
        
        # Print features
        print("FEATURES:")
        print(features_df)
        print("\nTARGET LABELS:")
        print(targets_df)
        print("=" * 80)
        
        # Map numerical targets back to original class labels
        original_labels = []
        for i in range(len(samples)):
            y_vec = y_sample[i]
            if np.sum(y_vec) > 0:  # Check if we have a valid one-hot vector
                class_idx = np.argmax(y_vec)
                class_name = self.y_train_dummies.columns[class_idx]
                original_labels.append(class_name)
            else:
                original_labels.append("unknown")
        
        print("\nOriginal class labels:")
        for i, label in enumerate(original_labels):
            print(f"Sample {i}: {label}")
        
        # Also show original features before encoding/scaling for context
        print("\nOriginal samples (before encoding/scaling):")
        print(self.raw_data.head(n))
    
    def get_data(self):
        """
        Get all the processed data.
        
        :return: (x_train_np, y_train_np), (x_test_np, y_test_np), feature_names, 
                 social_feature_indices, (train_dataset, test_dataset)
        """
        return ((self.x_train_np, self.y_train_np), 
                (self.x_test_np, self.y_test_np), 
                self.feature_names, 
                self.social_feature_indices, 
                (self.train_dataset, self.test_dataset))
```

```python
# Initialize and load the dataset
nursery = NurseryDataset(test_set=0.5, transform_social=True)

# Print dataset information
print(f"Training data shape: {nursery.x_train_np.shape}")
print(f"Testing data shape: {nursery.x_test_np.shape}")
print(f"Social feature indices: {nursery.social_feature_indices}")

# Print the head of the dataset with features and target labels
nursery.print_dataset_head(n=3)

# If you need the raw data components
(x_train_np, y_train_np), (x_test_np, y_test_np), feature_names, social_indices, (train_dataset, test_dataset) = nursery.get_data()
```

# Simple MLP Classifier


```python
# %% Cell 5 - Define PyTorch model for classification
class NurseryClassifier(nn.Module):
    def __init__(self, input_size, hidden_size, num_classes):
        super(NurseryClassifier, self).__init__()
        self.layer1 = nn.Linear(input_size, hidden_size)
        self.relu = nn.ReLU()
        self.layer2 = nn.Linear(hidden_size, num_classes)
        self.softmax = nn.Softmax(dim=1)
        
    def forward(self, x):
        x = self.layer1(x)
        x = self.relu(x)
        x = self.layer2(x)
        x = self.softmax(x)
        return x
    
    def predict_proba(self, x):
        """Method to match scikit-learn's API"""
        with torch.no_grad():
            if isinstance(x, np.ndarray):
                x = torch.FloatTensor(x)
            return self.forward(x).numpy()
```

```python

# %% Cell 6 - Train the PyTorch model
def train_model(model, train_dataset, criterion, optimizer, num_epochs=10, batch_size=64):
    train_loader = DataLoader(dataset=train_dataset, batch_size=batch_size, shuffle=True)
    
    # Training loop
    for epoch in range(num_epochs):
        total_loss = 0
        for inputs, labels in train_loader:
            # Forward pass
            outputs = model(inputs)
            loss = criterion(outputs, labels.float())
            
            # Backward and optimize
            optimizer.zero_grad()
            loss.backward()
            optimizer.step()
            
            total_loss += loss.item()
        
        # Print epoch statistics
        if (epoch+1) % 2 == 0:
            print(f'Epoch [{epoch+1}/{num_epochs}], Loss: {total_loss/len(train_loader):.4f}')
    
    return model

# Define model parameters
input_size = x_train_np.shape[1]  # Number of features
hidden_size = 64
num_classes = len(np.unique(y_train_np))

# Initialize the model
model = NurseryClassifier(input_size, hidden_size, num_classes)

# Define loss function and optimizer
criterion = nn.BCELoss()  # Binary Cross Entropy Loss
optimizer = optim.Adam(model.parameters(), lr=0.001)

# Train the model
model = train_model(model, train_dataset, criterion, optimizer, num_epochs=10)

# Evaluate the model
def evaluate_model(model, test_dataset):
    test_loader = DataLoader(dataset=test_dataset, batch_size=64, shuffle=False)
    correct = 0
    total = 0
    
    with torch.no_grad():
        for inputs, labels in test_loader:
            outputs = model(inputs)
            _, predicted = torch.max(outputs.data, 1)
            _, labels_max = torch.max(labels.data, 1)
            total += labels.size(0)
            correct += (predicted == labels_max).sum().item()
    
    accuracy = correct / total
    return accuracy

accuracy = evaluate_model(model, test_dataset)
print(f'Base model accuracy: {accuracy:.4f}')
```

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

# Create a confusion matrix for the model predictions
def create_confusion_matrix(model, test_dataset):
    test_loader = DataLoader(dataset=test_dataset, batch_size=64, shuffle=False)
    all_predictions = []
    all_labels = []
    
    with torch.no_grad():
        for inputs, labels in test_loader:
            outputs = model(inputs)
            _, predicted = torch.max(outputs.data, 1)
            _, labels_max = torch.max(labels.data, 1)
            
            all_predictions.extend(predicted.cpu().numpy())
            all_labels.extend(labels_max.cpu().numpy())
    
    return all_predictions, all_labels

# Get predictions and true labels
predictions, true_labels = create_confusion_matrix(model, test_dataset)

# Get unique class names from the dataset
unique_classes = np.unique(true_labels)
class_names = [f'Class {i}' for i in unique_classes]

# Create confusion matrix
conf_matrix = pd.DataFrame(
    np.zeros((len(unique_classes), len(unique_classes))),
    index=class_names,
    columns=class_names
)

# Fill confusion matrix
for pred, true in zip(predictions, true_labels):
    conf_matrix.iloc[true, pred] += 1

# Normalize confusion matrix
norm_conf_matrix = conf_matrix.div(conf_matrix.sum(axis=1), axis=0)

# Plot confusion matrix
plt.figure(figsize=(10, 8))
sns.set(font_scale=1.2)
heatmap = sns.heatmap(norm_conf_matrix, annot=True, fmt='.2f', cmap='Blues', 
                     cbar_kws={'label': 'Normalized Frequency'})
heatmap.set_title('Normalized Confusion Matrix')
heatmap.set_xlabel('Predicted Class')
heatmap.set_ylabel('True Class')
plt.tight_layout()
plt.show()

# Create a bar plot of class-wise accuracy
class_accuracy = np.zeros(len(unique_classes))
for i, class_idx in enumerate(unique_classes):
    # Get indices where true label is the current class
    indices = np.where(np.array(true_labels) == class_idx)[0]
    # Calculate accuracy for this class
    correct = np.sum(np.array(predictions)[indices] == class_idx)
    class_accuracy[i] = correct / len(indices) if len(indices) > 0 else 0

# Create a DataFrame for the class-wise accuracy
accuracy_df = pd.DataFrame({
    'Class': class_names,
    'Accuracy': class_accuracy
})

# Plot class-wise accuracy
plt.figure(figsize=(10, 6))
sns.set_style('whitegrid')
bar_plot = sns.barplot(x='Class', y='Accuracy', data=accuracy_df, palette='viridis')
bar_plot.set_title('Class-wise Accuracy')
bar_plot.set_ylim(0, 1)

# Add text labels on top of bars
for i, v in enumerate(class_accuracy):
    bar_plot.text(i, v + 0.02, f'{v:.4f}', ha='center')

plt.tight_layout()
plt.show()
```

the attack assumes an adversary has access to:

a trained target model
data samples missing a specific feature (the "attack feature")
predictions from the target model on these samples

in essence, it tries to infer/reconstruct private data that should have been hidden.
the main components:

the __init__ method sets up the attack by specifying:

the target model to attack
which feature to try to infer (via index)


the fit method trains an attack model:

it uses data where the sensitive feature is present to train
depending on how many unique values the feature has, it either:

builds a classification model (if ≤5 unique values)
builds a regression model (if >5 unique values)


the attack model takes as input:

all non-sensitive features
the target model's predictions




the infer method:

takes data WITHOUT the sensitive feature
plus predictions from the target model
runs them through the attack model
outputs the inferred values of the sensitive feature


helper classes:

AttackClassifier: simple neural network for classification attacks
AttackRegressor: similar network for regression attacks



this is essentially a demonstration of how a privacy leak might occur - where someone with access to a model's predictions could potentially recover sensitive information that was supposedly hidden from the inputs.

```python
# %% Cell 7 - Implement Black-box attack using PyTorch
class AttributeInferenceBlackBox:
    def __init__(self, target_model, attack_feature):
        """
        Initialize the black-box attack.
        
        :param target_model: The target model to attack
        :param attack_feature: Index of the feature to be attacked
        """
        self.target_model = target_model
        self.attack_feature = attack_feature
        self.attack_model = None
        self.unique_values = None
        
    def fit(self, x):
        """
        Train the attack model.
        
        :param x: Training data including the attack feature
        """
        # Get the values of the attack feature
        attack_feature_values = x[:, self.attack_feature].reshape(-1, 1)
        
        # Store unique values for later use
        self.unique_values = np.unique(attack_feature_values)
        print(f"Unique values in training: {self.unique_values}")
        
        # Get predictions from the target model using the full feature set
        x_tensor = torch.FloatTensor(x)
        with torch.no_grad():
            predictions = self.target_model(x_tensor).numpy()
        
        # Remove the attack feature from the training data
        x_without_feature = np.delete(x, self.attack_feature, axis=1)
        
        # Combine the remaining features with the predictions
        attack_train_data = np.concatenate((x_without_feature, predictions), axis=1)
        
        # Define a PyTorch model for the attack
        input_size = attack_train_data.shape[1]
        
        # If there are only a few unique values, we can treat it as a classification problem
        if len(self.unique_values) <= 5:  # Arbitrary threshold
            # Convert to class indices for classification
            attack_feature_classes = np.zeros(len(attack_feature_values), dtype=int)
            for i, val in enumerate(self.unique_values):
                attack_feature_classes[attack_feature_values.ravel() == val] = i
            
            # Create a classification model
            self.attack_model = AttackClassifier(input_size, 32, len(self.unique_values))
            
            # Convert data to PyTorch tensors
            attack_data_tensor = torch.FloatTensor(attack_train_data)
            attack_labels_tensor = torch.LongTensor(attack_feature_classes)
            
            # Train the attack model
            self._train_classifier(attack_data_tensor, attack_labels_tensor)
            print("Using classification model for attack")
        else:
            # Create a regression model
            self.attack_model = AttackRegressor(input_size, 32, 1)
            
            # Convert data to PyTorch tensors
            attack_data_tensor = torch.FloatTensor(attack_train_data)
            attack_labels_tensor = torch.FloatTensor(attack_feature_values)
            
            # Train the attack model
            self._train_regressor(attack_data_tensor, attack_labels_tensor)
            print("Using regression model for attack")
    
    def _train_classifier(self, x_tensor, y_tensor, epochs=50, batch_size=32):
        """
        Train the attack classifier model.
        
        :param x_tensor: Input tensor
        :param y_tensor: Target tensor
        :param epochs: Number of training epochs
        :param batch_size: Batch size for training
        """
        # Create dataset and dataloader
        dataset = TensorDataset(x_tensor, y_tensor)
        dataloader = DataLoader(dataset, batch_size=batch_size, shuffle=True)
        
        # Define loss function and optimizer
        criterion = nn.CrossEntropyLoss()
        optimizer = optim.Adam(self.attack_model.parameters(), lr=0.001)
        
        # Training loop
        for epoch in range(epochs):
            total_loss = 0
            for inputs, labels in dataloader:
                # Forward pass
                outputs = self.attack_model(inputs)
                loss = criterion(outputs, labels)
                
                # Backward and optimize
                optimizer.zero_grad()
                loss.backward()
                optimizer.step()
                
                total_loss += loss.item()
            
            # Print epoch statistics
            if (epoch+1) % 10 == 0:
                print(f'Attack model - Epoch [{epoch+1}/{epochs}], Loss: {total_loss/len(dataloader):.4f}')
    
    def _train_regressor(self, x_tensor, y_tensor, epochs=50, batch_size=32):
        """
        Train the attack regressor model.
        
        :param x_tensor: Input tensor
        :param y_tensor: Target tensor
        :param epochs: Number of training epochs
        :param batch_size: Batch size for training
        """
        # Create dataset and dataloader
        dataset = TensorDataset(x_tensor, y_tensor)
        dataloader = DataLoader(dataset, batch_size=batch_size, shuffle=True)
        
        # Define loss function and optimizer
        criterion = nn.MSELoss()
        optimizer = optim.Adam(self.attack_model.parameters(), lr=0.001)
        
        # Training loop
        for epoch in range(epochs):
            total_loss = 0
            for inputs, labels in dataloader:
                # Forward pass
                outputs = self.attack_model(inputs)
                loss = criterion(outputs, labels)
                
                # Backward and optimize
                optimizer.zero_grad()
                loss.backward()
                optimizer.step()
                
                total_loss += loss.item()
            
            # Print epoch statistics
            if (epoch+1) % 10 == 0:
                print(f'Attack model - Epoch [{epoch+1}/{epochs}], Loss: {total_loss/len(dataloader):.4f}')
    
    def infer(self, x, pred=None, values=None):
        """
        Infer the attack feature values.
        
        :param x: Data without the attack feature
        :param pred: Predictions from the target model (optional)
        :param values: Possible values of the attack feature
        :return: Inferred values of the attack feature
        """
        if pred is None:
            # We would need the full data with all features to get predictions
            # This would not be available in a real attack scenario
            raise ValueError("Predictions must be provided when the feature is removed")
        
        # Combine the features with the predictions
        attack_test_data = np.concatenate((x, pred), axis=1)
        attack_test_tensor = torch.FloatTensor(attack_test_data)
        
        # Get predictions from the attack model
        with torch.no_grad():
            raw_predictions = self.attack_model(attack_test_tensor).numpy()
        
        # Process the predictions based on the model type
        if isinstance(self.attack_model, AttackClassifier):
            # For classification, get the class with highest probability
            class_indices = np.argmax(raw_predictions, axis=1)
            
            # Map class indices back to actual values
            inferred_values = np.zeros(len(class_indices))
            for i, idx in enumerate(class_indices):
                inferred_values[i] = self.unique_values[idx]
        else:
            # For regression, use the raw predictions
            inferred_values = raw_predictions.ravel()
        
        return inferred_values.reshape(1, -1)

# Define the attack models
class AttackClassifier(nn.Module):
    def __init__(self, input_size, hidden_size, num_classes):
        super(AttackClassifier, self).__init__()
        self.layer1 = nn.Linear(input_size, hidden_size)
        self.relu = nn.ReLU()
        self.layer2 = nn.Linear(hidden_size, num_classes)
        
    def forward(self, x):
        x = self.layer1(x)
        x = self.relu(x)
        x = self.layer2(x)
        return x

class AttackRegressor(nn.Module):
    def __init__(self, input_size, hidden_size, output_size=1):
        super(AttackRegressor, self).__init__()
        self.layer1 = nn.Linear(input_size, hidden_size)
        self.relu = nn.ReLU()
        self.layer2 = nn.Linear(hidden_size, output_size)
        
    def forward(self, x):
        x = self.layer1(x)
        x = self.relu(x)
        x = self.layer2(x)
        return x
```

```python
# %% Cell 8 - Execute Black-box attack
# Split the training data for the attack
attack_train_ratio = 0.5
attack_train_size = int(len(x_train_np) * attack_train_ratio)
attack_test_size = int(len(x_train_np) * attack_train_ratio)
attack_x_train = x_train_np[:attack_train_size]
attack_y_train = y_train_np[:attack_train_size]
attack_x_test = x_train_np[attack_train_size:]
attack_y_test = y_train_np[attack_train_size:]

# Define the attack feature (social feature)
# We'll use the first social feature index from our preprocessing
attack_feature = social_indices[0]
print(f"Attacking feature at index: {attack_feature}")

# For the model predictions, we need to use the full feature set
# since the model was trained on all features
x_test_tensor = torch.FloatTensor(attack_x_test)
with torch.no_grad():
    attack_x_test_predictions = model(x_test_tensor).numpy()

# Then remove the attack feature for the attack model
attack_x_test_without_feature = np.delete(attack_x_test, attack_feature, 1)

# Extract the attacked feature for evaluation
attack_x_test_feature = attack_x_test[:, attack_feature].copy().reshape(-1, 1)

# Create and train the black-box attack
bb_attack = AttributeInferenceBlackBox(model, attack_feature=attack_feature)
bb_attack.fit(attack_x_train)
```

```python
# %% Cell 9 - Infer sensitive feature and check accuracy
# Get the unique values of the attacked feature
unique_values = np.unique(attack_x_test_feature)
print(f"Unique values of the attacked feature: {unique_values}")

# Infer the attacked feature values
inferred_train_bb = bb_attack.infer(attack_x_test_without_feature, pred=attack_x_test_predictions, values=unique_values)

# Check accuracy
train_acc = np.sum(inferred_train_bb == attack_x_test_feature.reshape(1, -1)) / len(inferred_train_bb.ravel())
print(f"Black-box attack accuracy: {train_acc:.4f}")

# %% Cell 10 - Implement White-box attack using PyTorch
class AttributeInferenceWhiteBox:
    def __init__(self, target_model, attack_feature):
        """
        Initialize the white-box attack.
        
        :param target_model: The target model to attack
        :param attack_feature: Index of the feature to be attacked
        """
        self.target_model = target_model
        self.attack_feature = attack_feature
        self.is_continuous = False
    
    def infer(self, x, pred=None, values=None, priors=None):
        """
        Infer the attack feature values.
        
        :param x: Data without the attack feature
        :param pred: Predictions from the target model (optional)
        :param values: Possible values of the attack feature
        :param priors: Prior probabilities of the feature values
        :return: Inferred values of the attack feature
        """
        if values is None:
            raise ValueError("Values of the attacked feature must be provided")
        
        # Check if we're dealing with continuous values
        if len(values) > 5 or np.issubdtype(values.dtype, np.floating):
            self.is_continuous = True
            print("Detected continuous values for white-box attack")
        else:
            print(f"Detected discrete values: {values}")
        
        if priors is None:
            # Use uniform priors if not provided
            priors = [1/len(values)] * len(values)
        
        # For white-box attacks, we need to analyze the model's behavior
        # when different values of the attack feature are used
        
        # Initialize the result array
        result = np.zeros(len(x))
        
        if self.is_continuous:
            # For continuous values, we'll use a simplified approach
            # We'll try a sample of values within the range
            min_val, max_val = np.min(values), np.max(values)
            test_values = np.linspace(min_val, max_val, min(10, len(values)))
            
            for i in range(len(x)):
                best_prob = -1
                best_value = test_values[0]
                
                for val in test_values:
                    # Create a copy of the sample
                    sample = x[i].copy()
                    
                    # Create a full-featured sample by inserting the current value at the attack feature position
                    full_sample = np.insert(sample, self.attack_feature, val)
                    
                    # Get the prediction probability for this reconstructed sample
                    try:
                        with torch.no_grad():
                            full_sample_tensor = torch.FloatTensor([full_sample])
                            pred_proba = self.target_model(full_sample_tensor).numpy()[0]
                        
                        prob = np.max(pred_proba)  # Use max probability
                        
                        if prob > best_prob:
                            best_prob = prob
                            best_value = val
                    except Exception as e:
                        # If prediction fails, try another value
                        print(f"Prediction failed for value {val}: {e}")
                        continue
                
                result[i] = best_value
        else:
            # For discrete values, try each possible value
            for i in range(len(x)):
                best_prob = -1
                best_value = values[0]
                
                for val_idx, val in enumerate(values):
                    # Create a copy of the sample
                    sample = x[i].copy()
                    
                    # Create a full-featured sample by inserting the current value at the attack feature position
                    full_sample = np.insert(sample, self.attack_feature, val)
                    
                    # Get the prediction probability for this reconstructed sample
                    try:
                        with torch.no_grad():
                            full_sample_tensor = torch.FloatTensor([full_sample])
                            pred_proba = self.target_model(full_sample_tensor).numpy()[0]
                        
                        prob = np.max(pred_proba) * priors[val_idx]  # Apply prior
                        
                        if prob > best_prob:
                            best_prob = prob
                            best_value = val
                    except Exception as e:
                        # If prediction fails, use the prior
                        print(f"Prediction failed for value {val}: {e}")
                        if val_idx < len(priors) and priors[val_idx] > best_prob:
                            best_prob = priors[val_idx]
                            best_value = val
                
                result[i] = best_value
        
        return result.reshape(1, -1)
```

```python
# %% Cell 11 - Execute White-box attack
# Create and execute the white-box attack
wb_attack = AttributeInferenceWhiteBox(model, attack_feature=attack_feature)

# Calculate priors based on the training data
unique_vals, counts = np.unique(attack_x_train[:, attack_feature], return_counts=True)
priors = counts / len(attack_x_train)
print(f"Unique values: {unique_vals}")
print(f"Prior probabilities: {priors}")

# Infer the attacked feature values
# For white-box attacks, we pass the data without the feature
inferred_train_wb = wb_attack.infer(attack_x_test_without_feature, 
                                  None,  # No need for predictions in white-box attack
                                  values=unique_values, 
                                  priors=priors)

# Check accuracy
train_acc_wb = np.sum(inferred_train_wb == attack_x_test_feature.reshape(1, -1)) / len(inferred_train_wb.ravel())
print(f"White-box attack accuracy: {train_acc_wb:.4f}")
```

```python
# %% Cell 12 - Implement baseline attack using PyTorch
class AttributeInferenceBaseline:
    def __init__(self, attack_feature):
        """
        Initialize the baseline attack.
        
        :param attack_feature: Index of the feature to be attacked
        """
        self.attack_feature = attack_feature
        self.attack_model = None
        self.unique_values = None
    
    def fit(self, x):
        """
        Train the baseline attack model.
        
        :param x: Training data including the attack feature
        """
        # Get the values of the attack feature
        attack_feature_values = x[:, self.attack_feature].reshape(-1, 1)
        
        # Store unique values for later use
        self.unique_values = np.unique(attack_feature_values)
        print(f"Baseline attack - Unique values in training: {self.unique_values}")
        
        # Remove the attack feature from the training data
        x_without_feature = np.delete(x, self.attack_feature, axis=1)
        
        # If there are only a few unique values, we can treat it as a classification problem
        if len(self.unique_values) <= 5:  # Arbitrary threshold
            # Convert to class indices for classification
            attack_feature_classes = np.zeros(len(attack_feature_values), dtype=int)
            for i, val in enumerate(self.unique_values):
                attack_feature_classes[attack_feature_values.ravel() == val] = i
            
            # Create a classification model
            input_size = x_without_feature.shape[1]
            self.attack_model = AttackClassifier(input_size, 32, len(self.unique_values))
            
            # Convert data to PyTorch tensors
            attack_data_tensor = torch.FloatTensor(x_without_feature)
            attack_labels_tensor = torch.LongTensor(attack_feature_classes)
            
            # Create dataset and dataloader
            dataset = TensorDataset(attack_data_tensor, attack_labels_tensor)
            dataloader = DataLoader(dataset, batch_size=32, shuffle=True)
            
            # Define loss function and optimizer
            criterion = nn.CrossEntropyLoss()
            optimizer = optim.Adam(self.attack_model.parameters(), lr=0.001)
            
            # Training loop
            for epoch in range(50):
                total_loss = 0
                for inputs, labels in dataloader:
                    # Forward pass
                    outputs = self.attack_model(inputs)
                    loss = criterion(outputs, labels)
                    
                    # Backward and optimize
                    optimizer.zero_grad()
                    loss.backward()
                    optimizer.step()
                    
                    total_loss += loss.item()
                
                # Print epoch statistics
                if (epoch+1) % 10 == 0:
                    print(f'Baseline - Epoch [{epoch+1}/50], Loss: {total_loss/len(dataloader):.4f}')
            
            print("Baseline using classification model")
        else:
            # Create a regression model
            input_size = x_without_feature.shape[1]
            self.attack_model = AttackRegressor(input_size, 32, 1)
            
            # Convert data to PyTorch tensors
            attack_data_tensor = torch.FloatTensor(x_without_feature)
            attack_labels_tensor = torch.FloatTensor(attack_feature_values)
            
            # Create dataset and dataloader
            dataset = TensorDataset(attack_data_tensor, attack_labels_tensor)
            dataloader = DataLoader(dataset, batch_size=32, shuffle=True)
            
            # Define loss function and optimizer
            criterion = nn.MSELoss()
            optimizer = optim.Adam(self.attack_model.parameters(), lr=0.001)
            
            # Training loop
            for epoch in range(50):
                total_loss = 0
                for inputs, labels in dataloader:
                    # Forward pass
                    outputs = self.attack_model(inputs)
                    loss = criterion(outputs, labels)
                    
                    # Backward and optimize
                    optimizer.zero_grad()
                    loss.backward()
                    optimizer.step()
                    
                    total_loss += loss.item()
                
                # Print epoch statistics
                if (epoch+1) % 10 == 0:
                    print(f'Baseline - Epoch [{epoch+1}/50], Loss: {total_loss/len(dataloader):.4f}')
            
            print("Baseline using regression model")
    
    def infer(self, x, values=None):
        """
        Infer the attack feature values.
        
        :param x: Data without the attack feature
        :param values: Possible values of the attack feature
        :return: Inferred values of the attack feature
        """
        # Convert input to tensor
        x_tensor = torch.FloatTensor(x)
        
        # Get predictions from the attack model
        with torch.no_grad():
            raw_predictions = self.attack_model(x_tensor).numpy()
        
        # Process the predictions based on the model type
        if isinstance(self.attack_model, AttackClassifier):
            # For classification, get the class with highest probability
            class_indices = np.argmax(raw_predictions, axis=1)
            
            # Map class indices back to actual values
            inferred_values = np.zeros(len(class_indices))
            for i, idx in enumerate(class_indices):
                inferred_values[i] = self.unique_values[idx]
        else:
            # For regression, use the raw predictions
            inferred_values = raw_predictions.ravel()
        
        return inferred_values.reshape(1, -1)
```

```python
# %% Cell 13 - Execute baseline attack
# Create and train the baseline attack
baseline_attack = AttributeInferenceBaseline(attack_feature=attack_feature)
baseline_attack.fit(attack_x_train)

# Infer the attacked feature values
inferred_train_baseline = baseline_attack.infer(attack_x_test_without_feature, values=unique_values)

# Check accuracy
baseline_train_acc = np.sum(inferred_train_baseline == attack_x_test_feature.reshape(1, -1)) / len(inferred_train_baseline.ravel())
print(f"Baseline attack accuracy: {baseline_train_acc:.4f}")
```

```python
# %% Cell 14 - Calculate precision and recall
def calc_precision_recall(predicted, actual, positive_value):
    """
    Calculate precision and recall for binary classification.
    
    :param predicted: Predicted values
    :param actual: Actual values
    :param positive_value: Value to consider as positive class
    :return: (precision, recall)
    """
    score = 0  # both predicted and actual are positive
    num_positive_predicted = 0  # predicted positive
    num_positive_actual = 0  # actual positive
    
    for i in range(len(predicted.ravel())):
        pred_val = predicted.ravel()[i]
        act_val = actual.ravel()[i]
        
        if pred_val == positive_value:
            num_positive_predicted += 1
        if act_val == positive_value:
            num_positive_actual += 1
        if pred_val == act_val and pred_val == positive_value:
            score += 1
    
    if num_positive_predicted == 0:
        precision = 1
    else:
        precision = score / num_positive_predicted
        
    if num_positive_actual == 0:
        recall = 1
    else:
        recall = score / num_positive_actual
    
    return precision, recall

# Calculate precision and recall for black-box attack
positive_value = unique_values[1] if len(unique_values) > 1 else unique_values[0]
bb_precision, bb_recall = calc_precision_recall(inferred_train_bb, attack_x_test_feature, positive_value)
print(f"Black-box attack - Precision: {bb_precision:.4f}, Recall: {bb_recall:.4f}")

# Calculate precision and recall for white-box attack
wb_precision, wb_recall = calc_precision_recall(inferred_train_wb, attack_x_test_feature, positive_value)
print(f"White-box attack - Precision: {wb_precision:.4f}, Recall: {wb_recall:.4f}")

# Calculate precision and recall for baseline attack
baseline_precision, baseline_recall = calc_precision_recall(inferred_train_baseline, attack_x_test_feature, positive_value)
print(f"Baseline attack - Precision: {baseline_precision:.4f}, Recall: {baseline_recall:.4f}")
```

```python

# %% Cell 17 - Visualizing how attribute inference attacks work
# This visualization demonstrates the core concept of attribute inference attacks

# 1. Feature importance visualization - shows which features leak information
def plot_feature_importance(attack_model, feature_names, attack_feature_idx):
    # For PyTorch models, we'll use a surrogate model to get feature importance
    from sklearn.ensemble import RandomForestRegressor
    from sklearn.inspection import permutation_importance
    
    # Create a small dataset without the sensitive feature
    X_sample = np.delete(attack_x_test[:100], attack_feature_idx, axis=1)
    y_sample = attack_x_test_feature[:100].ravel()
    
    # Get predictions from our attack model
    with torch.no_grad():
        X_tensor = torch.FloatTensor(X_sample)
        preds = attack_x_test_predictions[:100]
        X_with_preds = np.concatenate((X_sample, preds), axis=1)
    
    # Check if we need to discretize the target
    unique_values = np.unique(y_sample)
    print(f"Unique target values: {unique_values}")
    
    # Use a regression model instead of classification
    surrogate = RandomForestRegressor(n_estimators=50)
    surrogate.fit(X_with_preds, y_sample)
    
    # Get feature importance
    result = permutation_importance(surrogate, X_with_preds, y_sample, n_repeats=10, random_state=42)
    importance = result.importances_mean
    
    # Create feature names including model outputs
    remaining_features = np.delete(feature_names, attack_feature_idx)
    all_features = list(remaining_features) + [f'Model_Output_{i}' for i in range(preds.shape[1])]
    
    # Sort by importance
    indices = np.argsort(importance)[::-1]
    sorted_importance = importance[indices]
    sorted_features = [all_features[i] for i in indices]
    
    # Plot top 15 features
    plt.figure(figsize=(12, 8))
    sns.set_style('whitegrid')
    sns.barplot(x=sorted_importance[:15], y=sorted_features[:15], palette='viridis')
    plt.title('Feature Importance for Attribute Inference Attack', fontsize=16)
    plt.xlabel('Importance Score', fontsize=14)
    plt.ylabel('Features', fontsize=14)
    plt.tight_layout()
    plt.show()

# 2. Visualization of information leakage through model outputs
def visualize_information_leakage():
    # Get model outputs for samples with different sensitive attribute values
    sensitive_values = np.unique(attack_x_test_feature)
    
    # Get indices for each sensitive value
    indices_by_value = {}
    for val in sensitive_values:
        indices_by_value[val] = np.where(attack_x_test_feature.ravel() == val)[0]
    
    # Get model predictions for each group
    predictions_by_value = {}
    for val, indices in indices_by_value.items():
        if len(indices) > 0:
            # Get a sample of predictions for this value
            sample_indices = indices[:min(100, len(indices))]
            predictions_by_value[val] = attack_x_test_predictions[sample_indices]
    
    # Create a t-SNE visualization of model outputs colored by sensitive attribute
    from sklearn.manifold import TSNE
    
    # Combine all predictions
    all_preds = np.vstack([preds for preds in predictions_by_value.values()])
    all_values = np.concatenate([np.full(len(preds), val) for val, preds in predictions_by_value.items()])
    
    # Apply t-SNE
    tsne = TSNE(n_components=2, random_state=42)
    tsne_result = tsne.fit_transform(all_preds)
    
    # Create a DataFrame for plotting
    tsne_df = pd.DataFrame({
        'x': tsne_result[:, 0],
        'y': tsne_result[:, 1],
        'Sensitive Attribute': all_values
    })
    
    # Plot
    plt.figure(figsize=(12, 10))
    sns.scatterplot(x='x', y='y', hue='Sensitive Attribute', data=tsne_df, palette='Set1', s=100, alpha=0.7)
    plt.title('t-SNE Visualization of Model Outputs Colored by Sensitive Attribute', fontsize=16)
    plt.xlabel('t-SNE Dimension 1', fontsize=14)
    plt.ylabel('t-SNE Dimension 2', fontsize=14)
    plt.legend(title='Sensitive Attribute Value', title_fontsize=14, fontsize=12)
    plt.tight_layout()
    plt.show()
    
    # Create a violin plot to show distribution of prediction probabilities by sensitive attribute
    plt.figure(figsize=(14, 8))
    
    # Reshape predictions for plotting
    plot_data = []
    for val, preds in predictions_by_value.items():
        for i in range(preds.shape[1]):  # For each output class
            for prob in preds[:, i]:  # For each sample's probability
                plot_data.append({
                    'Sensitive Attribute': val,
                    'Output Class': f'Class {i}',
                    'Probability': prob
                })
    
    plot_df = pd.DataFrame(plot_data)
    
    # Create violin plot
    sns.violinplot(x='Output Class', y='Probability', hue='Sensitive Attribute', 
                  data=plot_df, palette='Set1', split=True)
    plt.title('Distribution of Model Output Probabilities by Sensitive Attribute', fontsize=16)
    plt.xlabel('Model Output Class', fontsize=14)
    plt.ylabel('Prediction Probability', fontsize=14)
    plt.legend(title='Sensitive Attribute Value', title_fontsize=14, fontsize=12)
    plt.tight_layout()
    plt.show()

# 3. Visualization of attack success by confidence
def visualize_attack_confidence():
    # For PyTorch models, we need a different approach to get confidence
    # We'll use the absolute difference between predictions and actual values as a proxy for confidence
    
    # Get the predictions from the black-box attack
    predictions = inferred_train_bb.ravel()
    actual = attack_x_test_feature.ravel()
    
    # Calculate the absolute difference (lower means more confident)
    abs_diff = np.abs(predictions - actual)
    
    # Normalize to [0, 1] range and invert so higher means more confident
    if np.max(abs_diff) > 0:
        confidence = 1 - (abs_diff / np.max(abs_diff))
    else:
        confidence = np.ones(len(abs_diff))
    
    # Check if predictions are correct
    correct = (inferred_train_bb.ravel() == attack_x_test_feature.ravel())
    
    # Create a DataFrame for plotting
    confidence_df = pd.DataFrame({
        'Confidence': confidence,
        'Correct': ['Correct' if c else 'Incorrect' for c in correct],
        'True Value': attack_x_test_feature.ravel(),
        'Predicted Value': inferred_train_bb.ravel()
    })
    
    # Plot confidence distribution by correctness
    plt.figure(figsize=(12, 8))
    sns.histplot(data=confidence_df, x='Confidence', hue='Correct', bins=20, 
                element='step', common_norm=False, palette={'Correct': 'green', 'Incorrect': 'red'})
    plt.title('Distribution of Attack Model Confidence by Prediction Correctness', fontsize=16)
    plt.xlabel('Attack Model Confidence', fontsize=14)
    plt.ylabel('Count', fontsize=14)
    plt.legend(title='Prediction', title_fontsize=14, fontsize=12)
    plt.tight_layout()
    plt.show()
    
    # Create a scatter plot of confidence vs. correctness
    plt.figure(figsize=(12, 8))
    jitter = np.random.normal(0, 0.05, size=len(correct))
    plt.scatter(confidence, correct.astype(int) + jitter, alpha=0.6, 
               c=confidence, cmap='viridis', s=80)
    plt.colorbar(label='Confidence')
    plt.title('Relationship Between Attack Confidence and Correctness', fontsize=16)
    plt.xlabel('Attack Model Confidence', fontsize=14)
    plt.ylabel('Prediction Correctness (0=Wrong, 1=Correct)', fontsize=14)
    plt.yticks([0, 1], ['Incorrect', 'Correct'])
    plt.tight_layout()
    plt.show()

# Run the visualizations
print("Visualizing feature importance for attribute inference...")
plot_feature_importance(bb_attack.attack_model, feature_names, attack_feature)

print("\nVisualizing information leakage through model outputs...")
visualize_information_leakage()

print("\nVisualizing attack confidence and success...")
visualize_attack_confidence()
```
